## 2. Salin file aplikasi & data ke folder kerja Colab
Sesuaikan `DRIVE_DIR` bila folder Drive-mu berbeda.


```python
import os, shutil
DRIVE_DIR = "/content/drive/MyDrive/skripsi_merek"

wajib = ["app_merek.py", "haystack_didaftar_fonID.csv"]
opsional = ["cc.id.100.bin"]

for f in wajib + opsional:
    src = os.path.join(DRIVE_DIR, f)
    if os.path.exists(src):
        shutil.copy(src, f); print("Disalin:", f, f"({os.path.getsize(f)/1e6:.1f} MB)")
    elif f in wajib:
        print("!! WAJIB TAPI TAK ADA:", src)
    else:
        print("(opsional) tidak ada:", f, "-> skor semantik akan 0")

print("\nIsi folder kerja:")
!ls -la
```
